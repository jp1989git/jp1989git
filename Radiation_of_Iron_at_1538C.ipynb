```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
sigma = 5.670374419e-8  # Stefan-Boltzmann constant
temperature = 1538 + 273.15  # Temperature in Kelvin
density = 7.6  # Density of iron at 1538°C (g/cm³)

# Mass range (in grams)
mass_range = np.linspace(0, 10000, 100)

# Calculate radius from mass and density
radius = ((3 * mass_range) / (4 * np.pi * density))**(1/3)

# Calculate surface area
surface_area = 4 * np.pi * radius**2

# Calculate power radiated using Stefan-Boltzmann law
power_radiated = sigma * surface_area * temperature**4

# Create the plot
fig, ax = plt.subplots()
ax.plot(mass_range, power_radiated, label="Power Radiated")

# Add interactive tooltips
def hover(event):
    if event.inaxes == ax:
        for line in ax.lines:
            xdata, ydata = line.get_data()
            ind = ax.get_cursor_data(event)[1]
            x, y = xdata[ind], ydata[ind]
            ax.format_coord = lambda x, y: f'x={x:.2f}, y={y:.2f}'
            ax.figure.canvas.draw()

fig.canvas.mpl_connect('motion_notify_event', hover)

# Customize the plot
plt.xlabel("Mass (g)")
plt.ylabel("Power Radiated (W)")
plt.title("Power Radiated by Iron Sphere at 1538°C")
plt.grid(True)
plt.legend()
plt.show()
```
